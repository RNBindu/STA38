Visualizing the quality of wines using PCA and t-SNE

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

# Load the data
data = pd.read_csv("wine.csv")

# Separate features and labels
features = data.drop(columns=["quality", "color"])

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features)

# Apply t-SNE for dimensionality reduction
tsne = TSNE(n_components=2, perplexity=30, n_iter=300)
tsne_result = tsne.fit_transform(scaled_features)

# Visualize Quality Ratings using PCA
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=data["quality"], cmap="coolwarm")
plt.title("Quality Ratings: PCA")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label="Quality Rating")

# Visualize Quality Ratings using t-SNE
plt.subplot(1, 2, 2)
plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c=data["quality"], cmap="coolwarm")
plt.title("Quality Ratings: t-SNE")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.colorbar(label="Quality Rating")

plt.tight_layout()
plt.show()
```

Visualizing the quality of wines using kmeans clustering with PCA and t-SNE

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the data
data = pd.read_csv("wine.csv")

# Separate features and labels
features = data.drop(columns=["quality", "color"])

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features)

# Apply t-SNE for dimensionality reduction
tsne = TSNE(n_components=2, perplexity=30, n_iter=300)
tsne_result = tsne.fit_transform(scaled_features)

# Apply K-Means clustering for quality prediction (PCA)
kmeans_quality_pca = KMeans(n_clusters=2, random_state=42)
kmeans_clusters_quality_pca = kmeans_quality_pca.fit_predict(pca_result)

# Apply K-Means clustering for quality prediction (t-SNE)
kmeans_quality_tsne = KMeans(n_clusters=2, random_state=42)
kmeans_clusters_quality_tsne = kmeans_quality_tsne.fit_predict(tsne_result)

# Visualize K-Means clustering results for quality prediction (PCA)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=data["quality"], cmap="coolwarm")
plt.title("K-Means Clustering: Quality Ratings: PCA")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label="Quality Rating")

# Visualize K-Means clustering results for quality prediction (t-SNE)
plt.subplot(1, 2, 2)
plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c=data["quality"], cmap="coolwarm")
plt.title("K-Means Clustering: Quality Ratings: t-SNE")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.colorbar(label="Quality Rating")

plt.tight_layout()
plt.show()

```

The plots for t-sne and pca with and without clusters are the same for wine quality. These can be because of the following points:

When PCA and t-SNE plots with and without K-Means clustering show similar patterns for quality, it suggests that the chemical properties might not strongly determine quality ratings.

The lack of well-defined clusters or clear separation in the plots indicates that the chemical properties might not be the sole or primary determinants of the quality ratings.

It's possible that other factors not considered in the dataset, such as winemaking techniques, aging processes, or external influences, play a more significant role in determining wine quality.

The complexity of factors contributing to quality ratings might not be effectively captured by these dimensionality reduction techniques alone. 

Exploring additional data or features can provide deeper insights into the complexity of factors influencing quality ratings.

Visualizing the differentiation between white and red wines using PCA and t-SNE

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler

# Load the data
data = pd.read_csv("wine.csv")

# Separate features and labels
features = data.drop(columns=["quality", "color"])

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features)

# Apply t-SNE for dimensionality reduction
tsne = TSNE(n_components=2, perplexity=30, n_iter=300)
tsne_result = tsne.fit_transform(scaled_features)

# Visualize White vs Red Wines using PCA
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=data["color"].map({"red": "r", "white": "b"}))
plt.title("White vs Red Wines: PCA")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")

# Visualize White vs Red Wines using t-SNE
plt.subplot(1, 2, 2)
plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c=data["color"].map({"red": "r", "white": "b"}))
plt.title("White vs Red Wines: t-SNE")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")

plt.tight_layout()
plt.show()
```

Visualizing the differentiation between white and red wines using K-Means clustering with both PCA and t-SNE

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the data
data = pd.read_csv("wine.csv")

# Separate features and labels
features = data.drop(columns=["quality", "color"])

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features)

# Apply t-SNE for dimensionality reduction
tsne = TSNE(n_components=2, perplexity=30, n_iter=300)
tsne_result = tsne.fit_transform(scaled_features)

# Apply K-Means clustering for color prediction (PCA)
kmeans_color_pca = KMeans(n_clusters=2, random_state=42)
kmeans_clusters_color_pca = kmeans_color_pca.fit_predict(pca_result)

# Apply K-Means clustering for color prediction (t-SNE)
kmeans_color_tsne = KMeans(n_clusters=2, random_state=42)
kmeans_clusters_color_tsne = kmeans_color_tsne.fit_predict(tsne_result)

# Visualize K-Means clustering results for color prediction (PCA)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=kmeans_clusters_color_pca, cmap="coolwarm")
plt.title("K-Means Clustering: White vs Red Wines (PCA)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")

# Visualize K-Means clustering results for color prediction (t-SNE)
plt.subplot(1, 2, 2)
plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c=kmeans_clusters_color_tsne, cmap="coolwarm")
plt.title("K-Means Clustering: White vs Red Wines (t-SNE)")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")

plt.tight_layout()
plt.show()
```

Both PCA and t-SNE were employed to reduce the dimensionality of the data for visualization. PCA focuses on capturing variance along principal components, while t-SNE emphasizes local structures and pairwise similarities.


When applying PCA and t-SNE without clustering, the differentiation between red and white wines was not distinctly evident, implying that the chemical properties alone might not naturally lead to clear separation based on wine color.

Incorporating K-Means clustering helped accentuate the differences between red and white wines in both PCA and t-SNE plots. This suggests that the clustering algorithm improved the separation, making it more visually discernible.

However, for quality ratings, applying PCA and t-SNE both with and without clustering didn't result in substantial differentiation. This could indicate that the chemical properties alone might not be the sole determinants of wine quality. The lack of evident clusters or patterns suggests that other factors might contribute more significantly to quality.

In conclusion, the unsupervised techniques employed in this analysis help shed light on the relationships between chemical properties, wine color, and quality. While clear differentiation is observed between red and white wines with clustering, the results suggest that other factors beyond chemical properties might play a more substantial role in determining wine quality.
